## 1.2 Create a three-dimensional vecor and plot it on a graph

```
#import Axes3d 
from mpl_toolkits.mplot3d import Axes3D

#Create a 3 dimensional vector
vector3d = [.3, .5, .7]

vector = np.array([[0, 0, 0, vector3d[0], vector3d[1], vector3d[2]]])

X, Y, Z, U, V, W = zip(*vector)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()

```

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```
from math import e, pi
print(e)
print(pi)
```

```
#Scale original vector to create new vectors
vector1 = vector2d
vector2 = np.multiply(5, vector2d)
vector3 = np.multiply(pi, vector2d)
vector4 = np.multiply(e * -1, vector2d)

#Plot the vectors
plt.arrow(0, 0, vector1[0], vector1[1], head_width=.05, head_length=.05, color='red')
plt.arrow(0, 0, vector2[0], vector2[1], head_width=.05, head_length=.05, color='green')
plt.arrow(0, 0, vector3[0], vector3[1], head_width=.05, head_length=.05, color='blue')
plt.arrow(0, 0, vector4[0], vector4[1], head_width=.05, head_length=.05, color='yellow')

plt.xlim(-2, 3)
plt.ylim(-2, 3)
plt.title('Scaling vectors')

plt.show()
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
#Declare the matrices
A = np.array([[1, 2], [3, 4], [5, 6]])
B = np.array([2, 4, 6])
C = np.array([[9, 6, 3], [4, 7, 11]])
D = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
E = np.array([[1, 3], [5, 7]])

#Perform calcs
CD = np.matmul(C, D)
AE = np.matmul(A, E)
BA = np.matmul(B, A)

print(CD.shape)
print(AE.shape)
print(BA)
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
#Create the matrices
H = np.array([[12, 11],
              [7, 10]])
J = np.array([[0, 1, 2],
              [7, 10, 4],
              [3, 2, 0]])

#Perform calcs and output
print(np.linalg.det(H))
print(np.linalg.det(J))
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
#Determine the inverse of the matrices
Hinv = np.linalg.inv(H)
Jinv = np.linalg.inv(J)

#print output
print(Hinv)
print(Jinv)
```
